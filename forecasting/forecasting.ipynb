```python
#| label: delhi-map4
#| echo: false
#| output: asis

import folium
import pandas as pd
import geopandas as gpd

shapefile_path = "data/shapefile/Districts.shp"
delhi = gpd.read_file(shapefile_path)

stations = pd.read_csv("data/dataset/22_station_fill_forward.csv")
stations = stations[['latitude', 'longitude', 'station']].drop_duplicates()
# stations["station"] = stations["station"].str.replace(", Delhi - DPCC", "", regex=False)

bounds = delhi.total_bounds
south_west = [bounds[1], bounds[0]]
north_east = [bounds[3], bounds[2]]

delhi_map = folium.Map(
    min_zoom=10,
    max_zoom=18
)

delhi_map.fit_bounds([south_west, north_east])

delhi_map.options['maxBounds'] = [south_west, north_east]

for index, row in stations.iterrows():
    folium.Marker(
        location=[row["latitude"], row["longitude"]],
        tooltip=row["station"],
        popup=f'<a href="station.html?station={row["station"]}" target="_blank">View Forecast</a>'
    ).add_to(delhi_map)

folium.GeoJson(delhi.__geo_interface__).add_to(delhi_map)

delhi_map
```
